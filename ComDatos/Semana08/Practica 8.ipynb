Acerca de Cualquier Onda:

$x(t)=A\cos(\omega t+\phi )$

donde:

- $x$ es la elongación o desplazamiento respecto al punto de equilibrio.
- $A$ es la amplitud del movimiento (elongación máxima).
- $\omega $ es la frecuencia angular
- $t$ es el tiempo.
- $\phi $ es la fase inicial e indica el estado de oscilación o vibración (o fase) en el instante t = 0 de la partícula que oscila.
Además, la frecuencia de oscilación puede escribirse como esto:

$ f={\frac {\omega }{2\pi }}={\frac {1}{2\pi }}{\sqrt {\frac {k}{m}}}$ y por lo tanto el periodo como $ T={\frac {1}{f}}={\frac {2\pi }{\omega }}=2\pi {\sqrt {\frac {m}{k}}}$

La velocidad y aceleración de la partícula pueden obtenerse derivando respecto del tiempo la expresión $ x(t)=A\cos(\omega t+\phi )$ $ x(t)=A\cos(\omega t+\phi )$ .

```python
def SIGNAL(A, F, T, OFFSET, Res = int(1e5), Xo = 0, Xf = 1):
    x = np.arange(Xo, Xf + (1e-16), (1 / Res)) * T
    y = A * np.cos(((F * (2 * np.pi)) * x) + (OFFSET * 2 * np.pi))
    
    return x, y
```

```python
x1,y1 = SIGNAL(2.5, 0.5, 5, 0.25)
x2,y2 = SIGNAL(1.5, 1, 5, 0.5)

plt.grid()

# plt.subplot(2,3,1)
plt.plot(x1,y1)
# plt.subplot(2,3,3)
plt.plot(x2,y2)

# plt.subplot(2,3,5)
plt.plot(x1,y1 + y2)
```

```python
X = x1
Y = y1 + y2
```

```python
plt.grid()
plt.plot(X,Y)
```

```python
def SAMPLING(X,Y, ResX = 10, ResY = 4):
    bits = 2 ** ResY
    segstride = 1 / ResX
    
    XOffset, XStrides = math.modf( (X.max() - X.min()) / segstride)
    if XStrides == 0.0:
        XStrides = XOffset
        XOffset = 0
    else:
        XOffset = 1
    XStride = (np.ceil(X.max()) - np.ceil(X.min())) / (XStrides + XOffset)

    YStride = (np.ceil(Y.max()) - np.ceil(Y.min())) / bits #(YStrides + YOffset)
    x = np.arange(np.ceil(X.min()), np.ceil(X.max()) + 1e-10, XStride)
    y = np.arange(np.ceil(Y.min()), np.ceil(Y.max()) + 1e-10, YStride)
    
    xstack = np.stack((x[:-1], x[1:]))
    ystack = np.stack((y[:-1], y[1:]))
    
    xmid = (x[:-1] + x[1:]) / 2
    ymid = (y[:-1] + y[1:]) / 2
    
    localY = np.zeros(xstack.shape[1])
    for ii in range(xstack.shape[1]):
        localrange = xstack[:, ii]
        localY[ii] = np.mean(Y[(X >= localrange[0]).astype(int) + (X <= localrange[1]).astype(int) == 2])
        
    yout = np.zeros(localY.shape)
    for jj in range(ystack.shape[1]):
        localrange = ystack[:, jj]
        yout[(localY >= localrange[0]).astype(int) + (localY <= localrange[1]).astype(int) == 2] = ymid[jj]
    
    yout = np.stack((yout, yout)).flatten('F')    
    xout = xstack.flatten('F')
    
    return xout, yout
```

```python
x,y = SAMPLING(X,Y)
plt.grid()
plt.plot(X,Y)
plt.plot(x,y)
```

La funcion SIGNAL esta compuesta por los parametros: Amplitud, Frecuencia, Tiempo y Fase Inicial.
- La amplitud es un Real Positivo mayor que 0,Define el Alto ( Eje Y ) de la onda
- Frecuencia es un Real Positivo mayor que 0, define el Ancho ( Eje X ) de la onda , midiendo la distancia entre cima y cima
- Tiempo es un Real Positivo mayor que 0, define el Ancho Total Maximo del Eje X
- Fase Inicial, es un Real cualesquiera, define el punto inicial (en el eje y) del cual comienza la onda

La funcion SAMPLING esta compuesta por los parametros: Valores originales X, Valores originales Y, Muestras por Segundo, Muestras por Amplitud total.

- Los parametros son autodescriptivos
    * las muestras por segundo, es un numero entero positivo mayor que 1, que define el numero de muestras por segundo no por toda la muestra
    * las muestras por amplitud es el numero de bits para el muestro del 'volumen'.

## EJemplo 1

```python
X,Y = SIGNAL(1, 0.5, 5, 0.25)

x,y = SAMPLING(X,Y, 10, 4)

plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
plt.grid()
plt.plot(X,Y)
plt.subplot(1,3,2)
plt.grid()
plt.plot(x,y)
plt.subplot(1,3,3)
plt.grid()
plt.plot(X,Y)
plt.plot(x,y)
```
